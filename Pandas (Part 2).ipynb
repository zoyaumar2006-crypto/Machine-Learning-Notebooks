## How to filter data in pandas

```python
import pandas as pd

dataset_sro = pd.read_csv('master.csv')
dataset_sro
```

```python
# to check a specific coloumn
dataset_sro.country
```

```python
dataset_sro.country.unique()
```

```python
dataset_sro[dataset_sro.country == 'Sri Lanka']
```

```python
# further filter to specific year
dataset_sro[(dataset_sro.country == 'Sri Lanka') & (dataset_sro.year == 1985)]
```

```python
cond1 = dataset_sro.country == 'Sri Lanka'
cond2 = dataset_sro.year == 1985
dataset_sro[cond1 & cond2]
```

```python
cond1 = dataset_sro.country == 'Sri Lanka'
cond2 = dataset_sro.year == 1985
cond3 = dataset_sro.year == 2006
dataset_sro[cond1 & (cond2 | cond3)]
```

```python
# to filter specific coloumns also
dataset_sro.loc[cond1 & (cond2 | cond3),['gender']]
```

```python
cond1 = dataset_sro.country == 'Sri Lanka'
cond2 = dataset_sro.year == 1985
cond3 = dataset_sro.year == 2006
cond4 = dataset_sro.gender == 'male'
cd = dataset_sro[cond1 & (cond2 | cond3) & cond4]
```

```python
# to sum the values of a coloumn
cd.sum().suicides_no
```

```python
cond1 = dataset_sro.country == 'Sri Lanka'
cond2 = dataset_sro.year >= 2000 #years greater or equal to 2000
cd = dataset_sro[cond1 & cond2]
cd
```

```python
cond1 = dataset_sro.country == 'Sri Lanka'
cond3 = dataset_sro.year.between(1990,2000) #data between these two years
cd = dataset_sro[cond1 & cond3]
cd
```

```python
cond1 = dataset_sro.country == 'Sri Lanka'
cond3 = dataset_sro.year.isin([1990,2000]) #data of these two years
cd = dataset_sro[cond1 & cond3]
cd
```

## Subplot
`(rows, columns, index)`   
Every time you call `plt.subplot(2, 2, 1)`, you are giving the computer three specific instructions:    
First Number (2): How many Rows does the grid have?    
Second Number (2): How many Columns does the grid have?      
Calculation: 2x2=4 total slots available.Third Number (1 to 4): Which Position are you currently drawing in?

```python
# subplot() is used to draw multiple plots
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)

plt.show()
```

```python
plt.figure(figsize = (10,10)) # to increase size

plt.subplot(2,2,1) #2x2 formation
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
# plt.subplot(2,2,5) #error (2x2=4)<5

plt.show()
```

```python
plt.figure(figsize = (10,10))

plt.subplot(3,2,1) #3x2 formation
plt.subplot(3,2,2)
plt.subplot(3,2,3)
plt.subplot(3,2,4)
plt.subplot(3,2,5)
plt.subplot(3,2,6)
# plt.subplot(3,2,7) #error (3x2=6)<7

plt.show()
```

```python
plt.figure(figsize = (10,10))

plt.subplot(321) #parameters can b without comma
plt.subplot(322)
plt.subplot(323)
plt.subplot(324)
plt.subplot(325)
plt.subplot(326)

plt.show()
```

```python
plt.figure(figsize = (10,10))

plt.subplot(2,2,1)
plt.pie([1])

plt.subplot(2,2,2)
plt.pie([1,2])

plt.subplot(2,2,3)
plt.pie([1,2,3])

plt.subplot(2,2,4)
plt.pie([1,2,3,4])

plt.show()
```
