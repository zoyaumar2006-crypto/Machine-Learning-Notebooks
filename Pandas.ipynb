### Interrogating the Data (Exploration)  
Once the data is loaded, you don't want to print all 5,000 rows. You use these keywords to "peek" inside.   
Keyword: **`df.head(n)`**  
What it does: Shows the first **n** rows (default is 5).   
The Logic: You use this to verify the columns loaded correctly and to see what the data actually looks like.  
Example: **`df.head(10)`** shows you the top 10 movies.  
Keyword: **`df.info()`**   
What it does: Shows the "Health Report" of the table.   
The Logic: It tells you:How many rows total (RangeIndex).Which columns have Missing Data (Non-Null Count).What Data Types are used (integers, floats, or "object"—which usually means text).   
Why it's vital for ML: If budget is stored as text (object) instead of a number (int), your model will crash. .info() helps you catch that.  
Keyword: **`df.describe()`**   
What it does: Gives you the "Quick Stats."   
The Logic: For every numerical column (like budget, revenue, runtime), it automatically calculates:   
Mean: The average.  
50% (Median): The middle value.   
Std: How spread out the data is.   
Example: Running **`df['runtime'].describe()`** tells you immediately if the average movie is 90 or 120 minutes long.   
Keyword: **`.shape`**   
`.shape[0]` and `.shape[1]`. In Python, .shape is an attribute (a property), not a function, which is why we don't use parentheses () after it.   
It returns a tuple: (Rows, Columns).    
`df.shape[0]`: Gives you the count of Rows (4803).   
`df.shape[1]`: Gives you the count of Columns (20).

### **`loc vs iloc`**

**Syntax:** Selecting Columns
Single Column: **df`['title']`** (Returns a Series).  
Multiple Columns: **`df[['title', 'budget']]`** (Returns a DataFrame—notice the double brackets).

```python
# Select the first 5 rows and the first 3 columns
top_rows_cols = df.iloc[0:5, 0:3]
print (top_rows_cols)
```

```python
# To use .loc effectively, it helps to set a text index
df_indexed = df.set_index('original_title')

# Now we can "search" for a specific movie by its name
avatar_data = df_indexed.loc['Avatar', ['budget', 'revenue']]
print (avatar_data)
```

```python
# --- ILOC (Integer-based) ---
# Get the first 3 movies and the first 2 columns
print(df.iloc[0:3, 0:2])

# --- LOC (Label-based) ---
# To use .loc effectively, let's set 'original_title' as the index
df_titles = df.set_index('original_title')

# Now we can "search" by movie name
print(df_titles.loc[['Avatar', 'Spectre'], ['budget', 'revenue']])
```

```python
# 1. Look at the data before (The index is just 0, 1, 2...)
print(df.head())

# 2. Change the 'address' system of the table
df_indexed = df.set_index('original_title')

# 3. Look at it now (The Movie Titles are now the bold 'address' on the left)
print(df_indexed.head())

# 4. Now use the Human-Friendly name to find data
# Syntax: .loc[Row_Name, Column_Name]
avatar_budget = df_indexed.loc['Avatar', 'budget']
print(f"The budget for Avatar was: {avatar_budget}")
```

## Phase 3: Data Cleaning (Preparing for ML)
Raw data is usually "dirty"—it has missing values, duplicates, and unnecessary columns. If you feed "dirty" data into a Machine Learning model, the model will give you "dirty" (inaccurate) results.

### Handling Missing Data (The "Gaps")
`.info()` result, columns like homepage and tagline have many missing entries. In Pandas, these are called NaN (Not a Number) or Nulls.   
**Keyword: `df.isnull().sum()`**    
What it does: This is your "Missing Data Map." It counts exactly how many empty cells are in every column.    
Why we use it: To decide if a column is worth keeping or if we need to fix it.   
**Keyword: `df.dropna()`**    
What it does: Deletes rows or columns that contain missing values.   
The Logic: If a movie has no release_date, we can't use it for time analysis, so we might just delete that one row.   
Syntax: `df.dropna(subset=['release_date'])` (Only deletes rows missing the date).   
**Keyword: `df.fillna(value)`**   
What it does: Fills in the gaps with a specific value.   
The Logic: Instead of deleting a row, we fill the gap with th`e Mean or a placeholder like "Unknown."

```python
# Task: Find which columns have empty spots
print(df.isnull().sum())
```

```python
# Before: 4803 rows
# We tell Pandas: "Only delete the row if 'release_date' is NaN"
df.dropna(subset=['tagline'], inplace=True)

# After: 4802 rows (The TMDB dataset has exactly 1 movie missing a date!)
print(df.shape)
```

```python
# This is dangerous with the TMDB dataset!
# Because 'homepage' has 3000+ missing values, this will delete 3000+ movies.
df_limited = df.dropna()
print(df_limited.shape)
```

```python
# Replace every single missing gap in the entire table with "Unknown"
#df_filled = df.fillna("Unknown")
df_filled = df['runtime'].fillna("Unknown")
```

```python
# Fill missing "runtime" gaps with the number 0
df['runtime'] = df['runtime'].fillna(0)

# Fill missing "tagline" gaps with "No Tagline Available"
df['tagline'] = df['tagline'].fillna("No Tagline Available")
```

```python
# Fill missing runtimes with the average runtime of all movies
avg_runtime = df['runtime'].mean()
df['runtime'] = df['runtime'].fillna(avg_runtime)
```

### Cleaning the Structure
**Keyword: `df.drop()`**   
What it does: Removes entire columns or rows that you don't need for your analysis.   
The Logic: For an ML model predicting revenue, the homepage URL is useless. Dropping it makes the data lighter and faster.   
**Keyword: `df.duplicated().sum()`**
What it does: Checks if any movie appears twice in the list.   
The Logic: Duplicates can "cheat" your model by making it see the same data twice.   
Syntax: df.drop_duplicates() removes the copies.

```python
# 'axis=1' means we are dropping a column, not a row
df_cleaned = df.drop(['homepage', 'tagline', 'overview'], axis=1)
print(df_cleaned)
```

```python
# 1. Check for duplicates (Returns the total count of identical rows)
duplicate_count = df.duplicated().sum()
print(f"Total duplicate rows found: {duplicate_count}")

# 2. To see WHICH rows are duplicated (optional but helpful for students)
# This shows only the rows that are copies
print(df[df.duplicated()])

# 3. Remove the duplicates from RAM
# Logic: This keeps the FIRST occurrence and deletes the rest.
df.drop_duplicates()

# 4. Verify
print(f"New shape after removing duplicates: {df.shape}")
```

## Phase 4: Feature Engineering
Now that we have cleaned the "gaps" and the "copies," we are in the most creative part of data science. This is where we prepare the data for the actual Machine Learning algorithms.

### Transforming Values with .apply()
The Logic: Sometimes you need a custom rule.    
For example, your students might want to create a new column called is_profitable (True/False) based on revenue and budget.

```python
# Define a custom rule
def check_profit(row):
    if row['revenue'] > row['budget']:
        return 'Profit'
    else:
        return 'Loss'

# Apply the rule to every row
# axis=1 means "apply this row-by-row"
df['profit_status'] = df.apply(check_profit, axis=1)

print(df[['original_title', 'profit_status']].head())
```

### One-Hot Encoding (Turning Text into Numbers)
The Logic: A model cannot do math on the word "Released" or "Rumored." We turn them into columns of 1s and 0s.   
**Keyword: `pd.get_dummies()`**
`pd.get_dummies(df['status'])`
This is the keyword that does the heavy lifting.  
It looks at the status column.   
It finds every unique word (Released, Rumored, etc.).  
It creates a brand new, temporary table (which you named status_encoded) where these words are now column headers.  
`pd.concat([df, status_encoded], axis=1)`   
`pd.concat`: This stands for concatenate (glue together).  
`[df, status_encoded]`: You are telling Python to take your original big table and your new "Encoded" table.   
axis=0 means glue them vertically (adding more rows).  
axis=1 means glue them horizontally (adding more columns).  

```python
# This turns the 'status' column into 'status_Released', 'status_Rumored', etc.
status_encoded = pd.get_dummies(df['status'])

# Join it back to our main table
df = pd.concat([df, status_encoded], axis=1)
```
