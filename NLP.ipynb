## Tokenization
Tokenization is the process of breaking down a large body of text into smaller units called tokens. This can be done at the sentence level or, more commonly, at the word level.

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# --- THIS IS THE KEY STEP ---
# You must download this 'resource' once so NLTK knows how to split text.
nltk.download('punkt_tab')

text = "Hello! Welcome to the NLP lab. We are learning preprocessing today."

# 1. Sentence Tokenization (Splitting by sentences)
sentences = sent_tokenize(text) 
print("Sentences:")
print(sentences)

# 2. Word Tokenization (Splitting by words/punctuation)
words = word_tokenize(text)
print("\nWords:")
print(words)
```

## Text Normalization (Lowercasing & Noise Removal)
Computers treat "Apple" and "apple" as two different entities because their ASCII values differ. To fix this, we convert everything to lowercase and remove "noise" like punctuation or special characters.   
1. The Function: `re.sub()`    
Purpose: sub stands for substitute.   
Logic: It looks for a specific pattern in your text and replaces it with something else.   
Structure: re.sub(pattern_to_find, what_to_replace_it_with, the_text_to_search)   
2. The Pattern: `r'[^\w\s]'   `
This is the most important part. It is the "filter" that identifies punctuation.   
r: This prefix stands for Raw String. It tells Python to read the symbols exactly as they are written without trying to interpret backslashes as special code.    
[ ]: These square brackets define a Set. Anything inside these brackets is what we are looking for.   
^: Inside brackets, this symbol means NOT. It flips the logic.    
\w: This is a shortcut for Alphanumeric characters (all letters a-z, A-Z, and numbers 0-9).    
\s: This is a shortcut for Whitespace (spaces, tabs, and newlines).    
The Translation: "Find every character that is NOT a letter, NOT a number, and NOT a space."     
Since punctuation marks (like !, ?, ., ,) are not letters, numbers, or spaces, they get "caught" by this filter.     
3. The Replacement: `''`   
This is an empty string.   
Because we told re.sub to replace the punctuation with "nothing," the punctuation simply disappears, effectively deleting it.

```python
import re

raw_text = "NLP is AWESOME!!!"

# 1. Lowercasing
clean_text = raw_text.lower()

# 2. Removing Punctuation using re.sub
# The pattern r'[^\w\s]' means: 
# Find everything that is NOT (^) a word character (\w) or whitespace (\s)
no_punc = re.sub(r'[^\w\s]', '', clean_text)

print(f"Original: {raw_text}")
print(f"No Punc:  {no_punc}")
# Output: nlp is awesome
```

## Stopword Removal    
Stopwords are common words that carry very little unique information, such as "the," "is," "at," and "which." Removing them reduces the "noise" in our data and allows the model to focus on the meaningful keywords.

```python
import nltk
from nltk.corpus import stopwords

# Ensure the stopwords are downloaded
nltk.download('stopwords')

# 1. Load the set of English stopwords
stop_words = set(stopwords.words('english'))

# 2. Your original list of words
words = ["this", "is", "a", "sample", "sentence", "about", "nlp"]

# 3. Perform the filtering
filtered_words = [w for w in words if w not in stop_words]
#filtered_words = [w for w in words if w.lower() not in stop_words]
# 4. Print outputs
print("--- Stopword Removal Analysis ---")
print(f"Original Words: {words}")
print(f"Stopwords found and removed: {[w for w in words if w in stop_words]}")
print(f"Final Filtered List: {filtered_words}")
```

## Stemming (The "Chop")
Stemming is a crude, rule-based process that chops off the ends of words. It doesn't care about grammar; it just follows a formula.   
Logic: If it ends in "ing," "ed," or "s," cut it off.  
Pros: Very fast.  
Cons: Can result in non-dictionary words (e.g., "university" becomes "univers")

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
words = ["running", "runner", "runs", "easily", "fairly"]

stems = [stemmer.stem(w) for w in words]
print(stems)
```

## Lemmatization (The "Dictionary")
Lemmatization is smarter. It uses a dictionary (WordNet) to find the Lemma (the actual root word).
Logic: Look up the word and its context to find the base form.   
Pros: Very accurate; always returns real words.   
Cons: Slower than stemming because it requires a dictionary look-up.   

```python
import nltk
from nltk.stem import WordNetLemmatizer

# You must download the dictionary first
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
words = ["running", "rocks", "better", "leaves"]

# 'better' becomes 'good' because lemmatizer knows the relationship
lemmas = [lemmatizer.lemmatize(w, pos="v") for w in words] # pos='v' tells it to treat as verbs
print(lemmas)
```
