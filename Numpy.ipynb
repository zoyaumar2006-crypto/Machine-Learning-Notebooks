## Why NumPy is Essential for Machine Learning
In Machine Learning, everything is a number. Whether you are dealing with text, images, or sensor data, it must be converted into a numerical format before a model can process it.
### 1. Data Representation
**Images:** A grayscale image is a 2D array of pixel intensities. A color image is a 3D array (Height, Width, RGB Channels).  
**Text:** Words are converted into numerical vectors (Embeddings).  
**Features:** A dataset with 1,000 rows and 10 features is simply a $1000 \times 10$ NumPy matrix.  
### 2. Mathematical Operations (Linear Algebra)
ML algorithms are essentially massive chains of Linear Algebra. NumPy handles:  
**Matrix Multiplication:** The "dot product" used in neural networks to calculate weights.  
**Transposition:** Flipping matrices to align dimensions for calculation.  
**Inversion & Eigenvalues:** Used in algorithms like Principal Component Analysis (PCA).
### 3. Efficiency & Speed
Training a model involves iterating over millions of data points. 
If you used standard Python loops, training a simple model would take hours; with NumPy’s C-optimized back-end, it takes seconds.

## View vs Copy

```python
import numpy as np

arr = np.array([5,10,15,20,25,30])
arr
```

```python
sliced_arr = arr[2:5] #just reflecting the view into variable
sliced_arr
```

```python
sliced_arr[:] = 0
```

```python
sliced_arr
```

```python
arr
```

```python
arr = np.array([5,10,15,20,25,30])
sliced_arr = arr[2:5].copy() #now storing copy into variable
sliced_arr[:] = 0
sliced_arr
```

```python
arr
```
