## 1. Line Plot
---
We use a line plot in Matplotlib to visualize trends or changes in data over a continuous interval, such as time.   
It helps in showing relationships between two variables clearly.

```python
# plot(x,y) function to plot a graph
plt.plot(days,temperature)
plt.show() #to show graph,not neccessary in jupyter
```

```python
plt.plot(days,temperature)
# axis([xmin, xmax, ymin, ymax])
plt.axis([0,20,30,50]) #to set axis
plt.title('Lahore\'s Temprature')
plt.xlabel('Days')
plt.ylabel('Temprature')
plt.show()
```

```python
# plt.plot(days,temperature, linestyle = ':') #change line style
# plt.plot(days,temperature, linestyle = '-')
plt.plot(days,temperature, color= 'r', marker = '>', markersize = 10 ,linestyle = '--')
plt.title('Lahore\'s Temprature')
plt.xlabel('Days')
plt.ylabel('Temprature')
plt.show()
```

```python
plt.plot(days,temperature, linestyle = '--', linewidth = 3) # to change line width
plt.title('Lahore\'s Temprature')
plt.xlabel('Days')
plt.ylabel('Temprature')
plt.show()
```

In Matplotlib, 'g>:' is a "Format String" (also known as a shorthand or "fmt" string). It is a quick, condensed way to tell Python three things at once without typing out every keyword.

Think of it like a secret code: `[Color][Marker][LineStyle]`.

```python
# plt.plot(days,temperature, 'ro--', linewidth = 2.5, markersize = 5) # 'colormarkerlinestyle' all in one place
plt.plot(days,temperature, 'g>:', linewidth = 2.5, markersize = 5)
# plt.plot(days,temperature, 'o--', linewidth = 2.5, markersize = 5)
plt.title('Lahore\'s Temprature')
plt.xlabel('Days')
plt.ylabel('Temprature')
plt.show()
```

```python
# use of legend to show line name
plt.plot(days,temperature, 'ro--', linewidth = 2.5, markersize = 5)
plt.title('Lahore\'s Temprature', fontsize=20)
plt.xlabel('Days', fontsize=16)
plt.ylabel('Temprature', fontsize=16)
plt.legend(['Temp line'], loc = 4) # to show line name
plt.show()
```

## 2. Histograms
---
Histograms are used to visualize the <b>distribution of a dataset</b> by <b>grouping data into bins</b> and showing how many values fall into each bin.   
It helps identify patterns like <b>skewness, spread, and outliers</b> in numerical data.   
**Use of `np.random.seed(12)`**   
Computers aren't actually "random"—they use complex mathematical formulas to create a sequence of numbers that look random. These formulas start with a "Seed" (a starting number).    
If you don't set a seed: Every time a student runs the code, they get different ages, different graphs, and different results.   
If you do set a seed: Every student in your lab will get the exact same "random" numbers as you.   
The number 12 itself isn't special. You could use 1, 42, or 2026.    
The Rule: As long as the seed number is the same, the "random" sequence will be identical every time.

```python
# used when we need to show info in bars
# to create data using random, we have to import numpy and random
import numpy as np
import random

np.random.seed(12)
ai_st_age = np.random.randint(18,45,100)
bd_st_age = np.random.randint(18,45,100)


# hist() is used to creat histogram
plt.hist(ai_st_age)
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] #used to make groups
plt.hist(ai_st_age, bins)
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15, 30,45]
plt.hist(ai_st_age, bins)
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ai_st_age, bins, rwidth=0.8) #to set width
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ai_st_age, bins, rwidth=0.8, histtype = 'step') # to change hist style
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()

plt.hist(ai_st_age, bins, rwidth=0.8, histtype = 'bar')
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()

plt.hist(ai_st_age, bins, rwidth=0.8, histtype = 'stepfilled')
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ai_st_age, bins, rwidth=0.8, orientation='horizontal') # to change orientation
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ai_st_age, bins, rwidth=0.8, color ='y') # to change color
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ai_st_age, bins, rwidth=0.8, color='r', label='AI students')
plt.hist(bd_st_age, bins, rwidth=0.8, color='g', label='BD students')
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist([ai_st_age, bd_st_age], bins, rwidth=0.8, color=['r','g'], label=['AI students', 'BD students'])
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,4)) # to change fig size 

bins = [15,20,25,30,35,40,45]
plt.hist([ai_st_age, bd_st_age], bins, rwidth=0.8, color=['r','g'], label=['AI students', 'BD students'])
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.legend()
plt.grid()
plt.show()
```

## Scatter plot
---
We use a scatter plot to display the relationship or correlation between two numerical variables.   
It helps identify patterns, trends, or outliers in data, especially for visualizing how one variable affects another.

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset_gpsd = pd.read_csv('googleplaystore.csv')
dataset_gpsd
```

```python
dataset_gpsd.shape
```

```python
dataset_gpsd = pd.read_csv('googleplaystore.csv', nrows = 500)
dataset_gpsd
dataset_gpsd.shape
```

```python
x = dataset_gpsd['Rating']
y = dataset_gpsd['Reviews']
```

```python
plt.scatter(x,y)
plt.title('Google Play Store Apps Scatter plot', fontsize = 12)
plt.xlabel('Rating', fontsize = 10)
plt.ylabel('Reviews', fontsize = 10)

plt.show()
```

```python
dataset_gpsd = pd.read_csv('googleplaystore.csv', nrows = 500)
dataset_gpsd
dataset_gpsd.shape

x = dataset_gpsd['Rating']
y = dataset_gpsd['Reviews']

plt.figure(figsize=(16,4))

plt.scatter(x,y, c = 'r', marker = '*', s = 100) #marker size change
plt.title('Google Play Store Apps Scatter plot', fontsize = 12)
plt.xlabel('Rating', fontsize = 10)
plt.ylabel('Reviews', fontsize = 10)
plt.show() 
```

## Pie chart

```python
# to show info in circular formate
import matplotlib.pyplot as plt

classes = ['Python','R','AI','ML','DS']
class1_students = [30,10,20,25,10]

plt.pie(class1_students)
plt.show()
```

```python
plt.pie(class1_students, labels = classes) #to give labels
plt.show()
```

```python
explode = [0,0.1,0,0,0.1]
plt.pie(class1_students, labels = classes, explode = explode ) #to slice a part
plt.show()
```

```python
colors = ['c','b','r','y','g']
plt.pie(class1_students, labels = classes, colors = colors) #to give colors
plt.show()
```

```python
plt.pie(class1_students, labels = classes, autopct ='%0.1f%%') #to show percentage
plt.show()
```

```python
plt.pie(class1_students, labels = classes, autopct ='%0.1f%%',
       radius = 2) #to change size
plt.show()
```

```python
plt.pie(class1_students, labels = classes, autopct ='%0.1f%%',
       radius = 1.5, startangle = 90) #to change size
plt.show()
```

```python
textprops = {'fontsize':15}
plt.pie(class1_students, labels = classes, autopct ='%0.1f%%',
       radius = 1.5, textprops = textprops) #to change text size
plt.show()
```
